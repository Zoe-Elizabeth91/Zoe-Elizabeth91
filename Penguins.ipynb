```python
import pandas as pd
import seaborn as sns

sns.set(rc={'figure.figsize':(10,8)})
penguins = pd.read_csv("penguins.csv")
penguins
```

```python
sns.set_style("darkgrid")
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
)
```

```python
sns.set_style("whitegrid")
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
)
```

```python
sns.set_style("ticks")
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
)
```

```python
sns.set_style("ticks")
```

```python
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
    palette="pastel"
)
```

```python
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
    palette="colorblind"
)
```

```python
ax = sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    hue="species",
    palette="colorblind"
)
ax.set_xlabel("Bill length (mm)")
ax.set_ylabel("Flipper length (mm)")          
ax.set_title("Penguins bill vs. Flipper length")
```
